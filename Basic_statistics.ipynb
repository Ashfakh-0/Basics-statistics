```python
#importing the libraries
import matplotlib.pyplot as plt
import seaborn as sns

# List of numerical columns
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value',
                  'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

for col in numerical_cols:
    plt.figure(figsize=(12,5))

    # Histogram
    plt.subplot(1,2,1)#(1,2,1)=(nrow,ncolumn,index(This is why histogram appears on the left and boxplot on the right))
    sns.histplot(data[col], bins=30, kde=True, color='skyblue')
    plt.title(f'Histogram of {col}')

    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(y=data[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')

    plt.tight_layout() #adjust the spacing
    plt.show()
```

```python
#here we can see most of the data are right skewed(+vely skewed = a few value rise the average)
#The histogram with two peaks indicate bimodality, meaning two common value ranges exist in the data
#and also we can see there is outliers in the data
```

```python
# skewness
for col in numerical_cols:
    skewness = data[col].skew()
    print(f"skewness of {col}:{skewness:.2f}")
```

```python
#here all the values are above 1 except skewness of discount rate, indicating datas are right skewed
#skewness of discount rate is -1.06. it's negatively skewed.it's mode is 5% so many discount is often occur at 5%.and we can see there is 2 peaks in histogram  indicate bimodality, meaning two common value ranges exist in the data
# here we can saw that there is outliers presents by looking in boxplot.
```
